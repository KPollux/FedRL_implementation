```python
maze_size = 8

def plot_rewards(episode_rewards, show_result=True, start_point=0, y_label='Rewards'):
    plt.figure(figsize=(10, 5))
    rewards_t = torch.tensor(episode_rewards, dtype=torch.float)
    if show_result:
        plt.title('Result')
    else:
        plt.clf()
        plt.title('Training...')
    
    
    
    plt.xlabel('Episode')
    plt.ylabel(y_label)
    plt.plot(rewards_t.numpy())
    plt.xlim(0, 500)
    # Take 100 episode averages and plot them too
    if len(rewards_t) >= 100:
        means = rewards_t.unfold(0, 100, 1).mean(1).view(-1)
        means = torch.cat((torch.zeros(99) + start_point, means))
        plt.plot(means.numpy())

    # plt.pause(0.001)  # pause a bit so that plots are updated
    # if is_ipython:
    #     if not show_result:
    #         display.display(plt.gcf())
    #         display.clear_output(wait=True)
    #     else:
    #         display.display(plt.gcf())
```

```python
plot_rewards(episode_rewards, True, -100, 'Rewards')
```

```python
plot_rewards(episode_step, True, 15, 'Steps')
```

```python
plot_rewards(win, True, 0, 'Win or not')
```
